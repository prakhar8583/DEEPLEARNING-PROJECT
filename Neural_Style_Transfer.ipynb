# Neural Style Transfer: Implementaion by using Tensorflow Hub

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_hub as hub
```

```python
# Next we Load The Image Lets Load The Cotent Images
# represt the content Image As C
C=plt.imread("/content/download (3).jfif")
```

```python
# Next We Load The style Images
S=plt.imread("/content/download (2).jfif")
```

```python
plt.imshow(S)
```

now we load the pretrained model from the tensorflow hub

```python
# now we load the pretrained model from the tensorflow hub
# Correct the typo in the hostname from thub.dev to tfhub.dev
model=hub.load("https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/1")
```

that model is available in tesorflow hub For Neural style transfer , Here Are several model is available , But this model is available in tensor flow hub

```python
# let Generate the resultant Images R using this tensorflowhub model
# BEfore That WE need To convert the style and content Images both the images as array of float value

```

```python
# let Generate the resultant Images R using this tensorflowhub model
# BEfore That WE need To convert the style and content Images both the images as array of float value
C=np.asarray(C,"float32")
S=np.asarray(S,"float32")
# Now the resultant Images as R

# Create TensorFlow constants from the numpy arrays, adding a batch dimension
content_image_tensor = tf.constant(C[np.newaxis, ...])
style_image_tensor = tf.constant(S[np.newaxis, ...])

# Call the model with the content and style image tensors
R = model(content_image_tensor, style_image_tensor)[0]
```

This Model Will generate the image in the form of tensor so we need to convert as numpy beacuse we using matplotlib as visuliaztions Task

```python
# lets convert to numpy array
R=np.array(R*255,dtype=np.uint8)
```
